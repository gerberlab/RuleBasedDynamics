```python
import matplotlib.pyplot as plt
from dynrules.data import get_cdiff_processed_logdata
import numpy as np
```

## load c. diff dataset

```python
t, xlog, mask_times, taxa = get_cdiff_processed_logdata(basepath="../CDIFF_DATA")
```

```python
num_time, num_subj, num_taxa = xlog.shape
```

```python
for subj in range(num_subj):
    fig, ax = plt.subplots()
    for oidx in range(num_taxa):
        ax.plot(t[1:], xlog[1:,subj,oidx], '-x', label=f'{taxa[oidx]}')
    ax.legend(bbox_to_anchor=(1.01,1.01))
    ax.set_xlabel("Time")
    ax.set_ylabel("Log abundance")
```
